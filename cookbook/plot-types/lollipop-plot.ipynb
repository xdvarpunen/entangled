# Lollipop Plot

## Lollipop Plot Quickstart

Lollipop plots can be created from a sequence of values

```
open ScottPlot

let myPlot = new Plot()

// Generate sine values
let values = Generate.Sin(25)

// Add lollipop plot
myPlot.Add.Lollipop(values)

myPlot
```

## Lollipop Positions

The position of each lollipop may be defined.

```
open ScottPlot

let myPlot = new Plot()

// Generate x-values
let xs = Generate.Range(0.0, 6.28, 0.314)

// Generate y-values as the sine of x-values
let ys = xs |> Array.map Math.Sin

// Add lollipop plot with specified x and y values
let lollipop = myPlot.Add.Lollipop(ys, xs)

myPlot
```

## Lollipop Plot Customizations

The stem line and head marker can be extensively customized.

```
open ScottPlot

let myPlot = new Plot()

// Generate sine values
let values = Generate.Sin(21)

// Add lollipop plot
let lollipop = myPlot.Add.Lollipop(values)

// Customize the lollipop markers
lollipop.MarkerColor <- Colors.Red
lollipop.MarkerSize <- 15f
lollipop.MarkerShape <- MarkerShape.FilledDiamond

// Customize the lollipop lines
lollipop.LineColor <- Colors.Green
lollipop.LineWidth <- 3f
lollipop.LinePattern <- LinePattern.Dotted

myPlot
```

## Horizontal Lollipop Plot

Change the lollipop plotâ€™s Orientation to Horizontal to cause stems to be drawn horizontally instead of vertically.

```
open ScottPlot

let myPlot = new Plot()

// Generate sine values for x-coordinates and consecutive numbers for y-coordinates
let xs = Generate.Sin(21)
let ys = Generate.Consecutive(21)
let coordinates = Coordinates.Zip(xs, ys)

// Add lollipop plot with horizontal orientation
let lollipop = myPlot.Add.Lollipop(coordinates)
lollipop.Orientation <- Orientation.Horizontal

myPlot
```
