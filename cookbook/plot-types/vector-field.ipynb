# Vector Field

## Vector Field Quickstart

Vectors (representing a magnitude and direction) can be placed at specific points in coordinate space to display as a vector field.

```
open ScottPlot
open System.Collections.Generic
open System.Numerics

// Create a new plot
let myPlot = Plot()

// Generate a grid of positions
let xs = Generate.Consecutive(10)
let ys = Generate.Consecutive(10)

// Create a collection of vectors
let vectors = List<RootedCoordinateVector>()

for i in 0 .. xs.Length - 1 do
    for j in 0 .. ys.Length - 1 do
        // Point on the grid
        let pt = Coordinates(xs[i], ys[j])

        // Direction & magnitude
        let dX = float32 ys[j]
        let dY = -9.81f / 0.5f * float32 (Math.Sin(xs[i]))
        let v = Vector2(dX, dY)

        // Add to the collection
        let vector = RootedCoordinateVector(pt, v)
        vectors.Add(vector)

// Plot the collection of rooted vectors as a vector field
myPlot.Add.VectorField(vectors)

myPlot
```

## Vector Field Colormap

Vector field arrows can be colored according to their magnitude.

```
open ScottPlot

// Create a new plot
let myPlot = new Plot()

// Generate sample vectors
let vectors = Generate.SampleVectors()

// Add a vector field to the plot
let vf = myPlot.Add.VectorField(vectors)

// Set the colormap
vf.Colormap <- ScottPlot.Colormaps.Turbo()

myPlot
```
