# Coxcomb Plot

## Coxcomb Plot Quickstart

A Coxcomb chart is a pie graph where the angle of slices is constant but the radii are not.

```
open System
open ScottPlot
open System.Collections.Generic

// Create a new plot
let myPlot = new Plot()

// Define the slices for the Coxcomb plot as IList<PieSlice>
let slices: IList<PieSlice> =
    List<PieSlice>([
        PieSlice(Value = 5.0, Label = "Red", FillColor = Colors.Red)
        PieSlice(Value = 2.0, Label = "Orange", FillColor = Colors.Orange)
        PieSlice(Value = 8.0, Label = "Gold", FillColor = Colors.Gold)
        PieSlice(Value = 4.0, Label = "Green", FillColor = Colors.Green.WithOpacity(0.5))
        PieSlice(Value = 8.0, Label = "Blue", FillColor = Colors.Blue.WithOpacity(0.5))
    ])

// Add the Coxcomb plot
myPlot.Add.Coxcomb(slices)

// Customize the plot
myPlot.Axes.Frameless()
myPlot.ShowLegend()
myPlot.HideGrid()

myPlot
```
