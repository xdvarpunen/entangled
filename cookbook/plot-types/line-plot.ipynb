# Line Plot

## Line Plot Quickstart

Line plots are placed with a start and end location in coordinate space. Their styles can be customized.

```
open ScottPlot

let myPlot = new Plot()

// Add lines to the plot
myPlot.Add.Line(1.0, 12.0, 12.0, 0.0)
myPlot.Add.Line(7.0, 9.0, 42.0, 9.0)
myPlot.Add.Line(30.0, 17.0, 30.0, 1.0)

myPlot
```

## Line Plot Shapes

Line plots can be styled using a LineStyle.

```
open ScottPlot

let myPlot = new Plot()
let colormap = new ScottPlot.Colormaps.Viridis()

let rand = new System.Random()

for i in 0 .. 9 do
    // Add a line with random start and end coordinates
    let lineStart = Generate.RandomCoordinates()
    let lineEnd = Generate.RandomCoordinates()
    let line = myPlot.Add.Line(lineStart, lineEnd)

    // Customize the line
    line.LineColor <- Generate.RandomColor(colormap)
    line.LineWidth <- float32(Generate.RandomInteger(1, 4))
    line.LinePattern <- Generate.RandomLinePattern()

    // Customize markers
    line.MarkerLineColor <- line.LineStyle.Color
    line.MarkerShape <- Generate.RandomMarkerShape()
    line.MarkerSize <- float32(Generate.RandomInteger(5, 15))

myPlot
```

## Line Plot Legend

Line plots with labels appear in the legend.

```
open ScottPlot

let myPlot = new Plot()

// Add sine signal
let sin = myPlot.Add.Signal(Generate.Sin())
sin.LegendText <- "Sine"

// Add cosine signal
let cos = myPlot.Add.Signal(Generate.Cos())
cos.LegendText <- "Cosine"

// Add a line plot
let line = myPlot.Add.Line(1.0, 12.0, 12.0, 0.0)
line.LineWidth <- 3f
line.MarkerSize <- 10f
line.LegendText <- "Line Plot"

// Show legend in the upper right corner
myPlot.ShowLegend(Alignment.UpperRight)

myPlot
```
