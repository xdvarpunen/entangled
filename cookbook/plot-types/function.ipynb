# Function

## Function Quickstart

Create a function plot from a formula.

```
open System
open ScottPlot

let myPlot = new Plot()

// Functions are defined as delegates with an input and output
let func1 (x: double) = Math.Sin(x) * Math.Sin(x / 2.0)
let func2 (x: double) = Math.Sin(x) * Math.Sin(x / 3.0)
let func3 (x: double) = Math.Cos(x) * Math.Sin(x / 5.0)

// Add functions to the plot
myPlot.Add.Function(func1)
myPlot.Add.Function(func2)
myPlot.Add.Function(func3)

// Manually set axis limits because functions do not have discrete data points
myPlot.Axes.SetLimits(-10.0, 10.0, -1.5, 1.5)

myPlot
```

## Function Limit X

A function can be limited to a range of X values.

```
open System
open ScottPlot

let myPlot = new Plot()

// Define the function
let func = Func<double, double>(fun x -> Math.Sin(x) * Math.Sin(x / 2.0))

// Add the function to the plot
let f = myPlot.Add.Function(func)
f.MinX <- -3.0
f.MaxX <- 3.0

// Set axis limits
myPlot.Axes.SetLimits(-5.0, 5.0, -0.2, 1.0)

myPlot
```
