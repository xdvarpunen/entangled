# Error Bars

## Error Bar Quickstart

Error Bars go well with scatter plots.

```
open ScottPlot

let myPlot = new Plot()

let points = 30

let xs = Generate.Consecutive(points)
let ys = Generate.RandomWalk(points)
let err = Generate.RandomSample(points, 0.1, 1.0)

let scatter = myPlot.Add.Scatter(xs, ys)
let errorbars = myPlot.Add.ErrorBar(xs, ys, err)
errorbars.Color <- scatter.Color

myPlot
```

## ErrorBar Values

Error size can be set for all dimensions.

```
open ScottPlot

let myPlot = new Plot()

let points = 10

let xs = Generate.Consecutive(points)
let ys = Generate.RandomWalk(points)

// Add scatter plot
let scatter = myPlot.Add.Scatter(xs, ys)
scatter.LineStyle.Width <- 0f // Hide the line

// Create error bars
let eb = ScottPlot.Plottables.ErrorBar(
    xs = xs,
    ys = ys,
    xErrorsNegative = Generate.RandomSample(points, 0.5),
    xErrorsPositive = Generate.RandomSample(points, 0.5),
    yErrorsNegative = Generate.RandomSample(points),
    yErrorsPositive = Generate.RandomSample(points)
)

eb.Color <- scatter.Color // Set color to match the scatter plot

myPlot.Add.Plottable(eb)

myPlot
```
