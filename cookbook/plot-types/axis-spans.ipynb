Axis Spans

## Axis Span Quickstart

Axis spans label a range of an axis. Vertical spans shade the full width of a vertical range, and horizontal spans shade the full height of a horizontal range.

```
open ScottPlot

// Create a new plot
let myPlot = new Plot()

// Add sine and cosine signals
myPlot.Add.Signal(Generate.Sin())
myPlot.Add.Signal(Generate.Cos())

// Add a horizontal span from y=10 to y=20
let hSpan = myPlot.Add.HorizontalSpan(10.0, 20.0)
hSpan.LegendText <- "Horizontal Span"

// Add a vertical span from x=0.25 to x=0.75
let vSpan = myPlot.Add.VerticalSpan(0.25, 0.75)
vSpan.LegendText <- "Vertical Span"

// Show the legend
myPlot.ShowLegend()

myPlot
```

## Axis Span Styling

Axis spans can be extensively customized.

```
open ScottPlot

// Create a new plot
let myPlot = new Plot()

// Add sine and cosine signals
myPlot.Add.Signal(Generate.Sin())
myPlot.Add.Signal(Generate.Cos())

// Add a horizontal span from y=10 to y=20
let hs = myPlot.Add.HorizontalSpan(10.0, 20.0)
hs.LegendText <- "Hello"

// Customize the line style of the horizontal span
hs.LineStyle.Width <- 2f
hs.LineStyle.Color <- Colors.Magenta
hs.LineStyle.Pattern <- LinePattern.Dashed

// Set the fill style of the horizontal span
hs.FillStyle.Color <- Colors.Magenta.WithAlpha(0.2)

// Show the legend
myPlot.ShowLegend()

myPlot
```
