# Phasor Plot

## Phasor Line Plot

A phasor line plot contains a collection of polar coordinates which are rendered as arrows.

```
open ScottPlot

let myPlot = new Plot()

// Start by placing a polar axis system on the plot
let polarAxis = myPlot.Add.PolarAxis(30)
polarAxis.Circles |> Seq.iter (fun x -> x.LinePattern <- LinePattern.Dotted)
polarAxis.Spokes |> Seq.iter (fun x -> x.LinePattern <- LinePattern.Dotted)

// A Phasor may be added with predefined points
let points1 = [
    PolarCoordinates(10.0, Angle.FromDegrees(15.0))
    PolarCoordinates(20.0, Angle.FromDegrees(120.0))
    PolarCoordinates(30.0, Angle.FromDegrees(240.0))
]
myPlot.Add.Phasor(points1)

// Points on a Phasor may be added or modified after it is created
let phasor2 = myPlot.Add.Phasor()
phasor2.Points.Add(PolarCoordinates(20.0, Angle.FromDegrees(35.0)))
phasor2.Points.Add(PolarCoordinates(25.0, Angle.FromDegrees(140.0)))
phasor2.Points.Add(PolarCoordinates(20.0, Angle.FromDegrees(260.0)))

myPlot
```

## Phasor Plot with Labels

Text labels may be applied to individual arrows of a phasor plot.

```
open ScottPlot

let myPlot = new Plot()

// Setup the polar axis
let polarAxis = myPlot.Add.PolarAxis(30)
polarAxis.Circles |> Seq.iter (fun x -> x.LinePattern <- LinePattern.Dotted)
polarAxis.Spokes |> Seq.iter (fun x -> x.LinePattern <- LinePattern.Dotted)

// Create a phasor plot and points in coordinate space
let phasor = myPlot.Add.Phasor()
phasor.Points.Add(PolarCoordinates(20.0, Angle.FromDegrees(35.0)))
phasor.Points.Add(PolarCoordinates(25.0, Angle.FromDegrees(140.0)))
phasor.Points.Add(PolarCoordinates(20.0, Angle.FromDegrees(260.0)))

// Add labels for points
phasor.Labels.Add("Alpha")
phasor.Labels.Add("Beta")
phasor.Labels.Add("Gamma")

// Style the labels
phasor.LabelStyle.FontSize <- 24f
phasor.LabelStyle.ForeColor <- Colors.Black
phasor.LabelStyle.FontName <- Fonts.Monospace
phasor.LabelStyle.Bold <- true

myPlot
```
