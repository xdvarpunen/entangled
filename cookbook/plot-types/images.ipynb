# Images

## Image Rectangle

An image can be drawn inside a rectangle defined in coordinate units.

```
open ScottPlot

let myPlot = new Plot()

// Load an image from the sample images
let img = ScottPlot.SampleImages.MonaLisa()

// Define the rectangle dimensions for the image
let rect = new CoordinateRect(left = 0.0, right = float img.Width, bottom = 0.0, top = float img.Height)

// Add the image rectangle to the plot
myPlot.Add.ImageRect(img, rect)

myPlot
```
