# Annotation

## Annotation Quickstart

Annotations are labels you can place on the data area of a plot. Unlike Text added to the plot (which is placed in coordinate units on the axes), Annotations are positioned relative to the data area (in pixel units) and do not move as the plot is panned and zoomed.

```
open ScottPlot

// Create a new plot
let myPlot = new Plot()

// Add sample data
myPlot.Add.Signal(Generate.Sin())
myPlot.Add.Signal(Generate.Cos())

// Add an annotation
myPlot.Add.Annotation("This is an Annotation")

myPlot
```

## Annotation Customization

Annotations can be extensively customized.

```
open ScottPlot

// Create a new plot
let myPlot = new Plot()

// Add sample data
myPlot.Add.Signal(Generate.Sin())
myPlot.Add.Signal(Generate.Cos())

// Add a customized annotation
let anno = myPlot.Add.Annotation("Customized\nAnnotation")
anno.LabelFontSize <- 32f
anno.LabelFontName <- Fonts.Serif
anno.LabelBackgroundColor <- Colors.RebeccaPurple.WithAlpha(0.3)
anno.LabelFontColor <- Colors.RebeccaPurple
anno.LabelBorderColor <- Colors.Green
anno.LabelBorderWidth <- 3f
anno.LabelShadowColor <- Colors.Transparent
anno.OffsetY <- 40f
anno.OffsetX <- 20f

myPlot
```

## Annotation Positions

Annotations are aligned with the data area.

```
open System
open ScottPlot

// Create a new plot
let myPlot = new Plot()

// Loop through each alignment value and add an annotation for each
for alignment in Enum.GetValues(typeof<Alignment>) do
    let anno = myPlot.Add.Annotation(alignment.ToString())
    anno.Alignment <- alignment :?> Alignment // Cast to Alignment type
    anno.OffsetX <- 0.0f // Adjust as needed
    anno.OffsetY <- 0.0f // Adjust as needed

// Set title and labels
myPlot.Title("Annotations for Each Alignment")
myPlot.XLabel("X Axis")
myPlot.YLabel("Y Axis")

myPlot
```
