# Callout

## Callout Quickstart

Callouts display a label and are connected with an arrow that marks a point on the plot.

```
open System
open ScottPlot

// Create a new plot
let myPlot = new Plot()

// Generate data
let xs = Generate.Consecutive(15)
let ys = Generate.Sin(15)
myPlot.Add.Scatter(xs, ys)

// Add callouts
myPlot.Add.Callout("Hello",
    textLocation = Coordinates(7.5, 0.8),
    tipLocation = Coordinates(xs.[6], ys.[6]))

myPlot.Add.Callout("World",
    textLocation = Coordinates(10.0, 0.0),
    tipLocation = Coordinates(xs.[13], ys.[13]))

myPlot
```
